```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("logs.csv")

df.head()
```

```python
clk1 = df.loc[df['clk'] == 1]
plt.hist(clk1['time'], range=(900, 1100), bins=50)
plt.ylabel("nr of samples")
plt.xlabel("time [us]")
plt.title("Jitter for 1000us clock")
```
