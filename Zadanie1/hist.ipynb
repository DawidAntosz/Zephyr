```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("logs.csv").tail(-3)

df.head()
```

```python
clk1 = df.loc[df['clk'] == 1]
plt.hist(clk1['period'], range=(900, 1100), bins=50)
plt.ylabel("nr of samples")
plt.xlabel("time [us]")
plt.title("Jitter for 1000us clock")
```

```python
plt.plot(clk1['time']/1e6, clk1['period'])
plt.xlabel("time since start [s]")
plt.ylabel("period [us]")
plt.title("Timer 1000us")
```
